# **Introduo**

**Tensorflow**  uma biblioteca de cdigo aberto do Google para machine learning e Deep Learning, permite criar e treinar modelos de IA, manipulando dados em tensores (array multidimensionais) e oferecendo ferramentas flexveis para pesquisadores e desenvolvedores, sendo usada em servios como Busca e Tradutor.  
  

  
## **Comandos** 

### **Input** 
Representa a entrada da rede neural.
Serve para dizer qual  o formato dos dados que vo entrar no modelo.  
 - Define o tamanho da imagem  
 -  sempre o primeiro bloco da rede
 - No tem pesos s descreve o formato

### **Flatten**
Transforma uma matriz (imagem) em um vetor nico. Imagens tm formato (altura, largura, canais)  
 necessrio quando voc vai conectar a imagem em camadas densas  
### **Camada Dense**

A camada **Dense** (totalmente conectada) conecta **todos os neurnios da camada atual a todos os neurnios da camada anterior**.  
 uma camada fundamental em redes neurais, usada para aprender padres complexos nos dados.

---

**Frmula**
  
`saida = activation((W * x) + b)`  
  

Onde:  
- **activation**  funo de ativao aplicada elemento a elemento  
- **W * x**  multiplicao matricial entre entrada e kernel (pesos)  
- **b**  vetor de bias  

---

  **Exemplo**

```python
densa = Dense(128, activation='relu')(x)
```
  
- Usada para classificao no final da rede  
- Pode ter ReLU, Softmax, Sigmoid etc.  
  
 **Caractersticas**

- Usada normalmente **no final da rede** para classificao  
- Aceita ativaes como **ReLU**, **Softmax**, **Sigmoid**, etc.

--

 **Parmetros**
```python
  dense = layers.Dense(
    units = 128,                           # *obrigatrio* Nmero de neurnios da camada
    activation = 'relu'                    # *opcional* Funo de ativao (`relu`, `sigmoid`, `softmax`, `tanh`, etc.).
    use_bias=True,                         # *(default: True)* Indica se um vetor de bias ser usado.
    kernel_initializer="glorot_uniform",   # Inicializao dos pesos (ex.: `glorot_uniform`, `he_normal`).
    bias_initializer="zeros",              # *(default: zeros)* Inicializao do bias.
    kernel_regularizer=None,               # Regularizao aplicada aos pesos (`l1`, `l2`, etc.).
    bias_regularizer=None,                 # Regularizao aplicada ao bias.
    activity_regularizer=None,             # Regularizao aplicada  sada da camada. 
    kernel_constraint=None,                # Restries aplicadas aos pesos (ex.: norma mxima).
    bias_constraint=None                   # Restries aplicadas ao bias (ex.: norma mxima).

  )           
```

### **Conv2D**  
 a camada convolucional, usada para identificar padres em imagens:  
- Arestas  
- Curvas  
- Texturas  
- Objetos  
Ela aplica **filtros** (kernels) pela imagem  
  
`conv = Conv2D(32, kernel_size=(3,3), activation='relu')(input)`  
  
Onde,  
- 32 filtros  
- Kernels 3x3  
- Cada filtro cria um mapa de caractersticas (feature map)  
  
### **Maxpooling**  
Reduz o tamanho da imagem pegando apenas o valor mximo dentro de janelas  
  
`pool = MaxPooling(pool_size=(2,2))(conv)`  
  
Isso diminui a resoluo, reduz parmetros, retm as caractersticas mais importantes, ajuda a evitar overfitting
